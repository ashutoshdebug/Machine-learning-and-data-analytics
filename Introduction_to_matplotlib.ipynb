```python
# Import matplotlib and get it ready for plotting in Jupyter
import matplotlib

# Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

# Setup plot
fig, ax=plt.subplots(figsize=(10,10))

# Plot data
ax.plot(x,y)

# Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# Save and show (You save the whole figure)
fig.savefig("sample-plot.png")
```

plt.subplot(figsize=()): figsize is parameter used to set the width and height of a figure in inches.<br>
ax.set(title="", xlabel="", ylabel=""):<br>
title="": Sets the title for axes.<br>
xlabel="": Sets the label for x-axis.<br>
ylabel="": Sets the label for y-axis.<br>

fig.savefig(): Saves the figure in the provided name and directory

# Making figures with NumPy Arrays
We want:
* Line Plot- Makes a single line plot.
* Scatter Plot- Plots a points on the axes.
* Bar Plot- Plots bar for each value on the axes.
* Histogram- Useful to show curves on the axes.

```python
import numpy as np
```

```python
# Create some data
x= np.linspace(0,10,100)
x
```

```python
# Plot the data and create a line plot
fig, ax= plt.subplots()
ax.plot(x, x**2);
```

```python
# Make a plot using a dictionary
nut_butter_price={"Almond Butter":10,
                 "Peanut Butter":8,
                 "Cashew Butter":12}
fig, ax= plt.subplots()
ax.set(title="Dan's Nut Butter Store",
      ylabel="Prices ($)")
ax.bar(nut_butter_price.keys(), nut_butter_price.values());
```

### Horizontal Bars

```python
fig, ax= plt.subplots()
ax.barh(list(nut_butter_price.keys()), list(nut_butter_price.values()));
```

By using a .barh() we can plot horizontal bar graphs

```python
# Make some data for histogram
x=np.random.randn(1000)
fig, ax= plt.subplots()
ax.hist(x);
```

## Two options for subplots

### First option:

```python
fig, ((ax1, ax2), (ax3, ax4))=plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.randn(1000));
```

### Second option:

```python
fig, ax =plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(10,5))
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_price.keys(), nut_butter_price.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method?)
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO mehthod

```python
over_50= heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
len(over_50)
```

```python
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO method
fig, ax= plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
over_50.target.values
```

```python
# OO method from the scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Targets");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True) # This will make the axes to share the same axis label

# Add data to axes
scatter= ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the ax0 plot:
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholesterol")

# Add a legend:
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# Add data to ax1:
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])

# Customize the ax1 plot:
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add a legend:
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line:
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title for figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1)=plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True)

# Add data to axes
scatter= ax0.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter")

# Change the x axis limits
ax0.set_xlim([50,80])
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Customize the ax0 plot:
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholesterol")

# Add a legend:
ax0.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# Add data to ax1:
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter")

# Customize the ax1 plot:
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# Add a legend:
ax1.legend(*scatter.legend_elements(), title="Target")

# Add a horizontal line:
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title for figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
