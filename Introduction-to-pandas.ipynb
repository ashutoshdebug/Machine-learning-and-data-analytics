# Shuffling the dataframe

We need to shuffle our dataframe to ensure that our machine learning model is capable of general dataframes

```python
car_sale_shuffle = car_sale.sample(frac=1)
car_sale_shuffle
```

As you can it changes the index of the data frame. 1 means all whole dataframe and 0.5 means half of dataframe in shuffle

```python
car_sale_shuffled=car_sale.sample(frac=0.5)
car_sale_shuffled
```

And our car_sale is intact:

```python
car_sale
```

```python
# Select and shuffles only 20% percent of the data:
car_sale_shuffle.sample(frac=0.2)
```

It can be used for much larger dataframes

Now if we want to reset our index we can use:

```python
car_sale_shuffle.reset_index()
```

As you can see we got another index column, we can resolve this by:

```python
car_sale_shuffle.reset_index(drop=True)
```

# .apply()
.apply() can be used to apply a function or do some sort of calculation and apply it all values of the column

```python
car_sale["Odometer (KM)"]=car_sale["Odometer (KM)"].apply(lambda x:x/1.6)
```

```python
car_sale
```

Here all the value of "Odometer (KM)" has been changed using the lambda function and converted into miles
